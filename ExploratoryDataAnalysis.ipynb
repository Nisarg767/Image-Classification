```python
#Data Analysis mnist
import numpy as np
import pandas as p
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
dt = p.read_csv('mnist_train.csv')
from scipy.linalg import eigh 
import matplotlib.pyplot as plt
import seaborn as sn

print(dt.head(5)) 

tg = dt['label']
tot = dt.drop("label",axis=1)
dataa = tot.head(15000)
sd_data = StandardScaler().fit_transform(dataa)
print(sd_data.shape)
sample = sd_data
cmatrix = np.matmul(sample.T , sample)

print ( "shape of variance matrix = ", cmatrix.shape)

v, vc = eigh(cmatrix, eigvals=(782,783))
lb = tg.head(15000)

print("Shape of eigen vectors = ",vc.shape)
vc = vc.T

print("Updated shape of eigen vectors = ",vc.shape)
nc = np.matmul(vc, sample.T)

print ("new data points' shape ", vc.shape, "X", sample.T.shape," = ", nc.shape)

nc = np.vstack((nc, lb)).T

dataf = p.DataFrame(data=nc, columns=("First_principal", "Second_principal", "label"))
print(dataf.head())
sn.FacetGrid(dataf, hue="label", size=6).map(plt.scatter, 'First_principal', 'Second_principal').add_legend()
plt.show()
```

```python
#cifar100
from sklearn.manifold import TSNE
from keras.datasets import mnist
from keras.datasets import cifar100
from sklearn.datasets import load_iris
from numpy import reshape
import seaborn as sns
import pandas as pd  
(x_train, y_train), (_ , _) = cifar100.load_data()
x_train = x_train[:3000]
y_train = y_train[:3000]
print(x_train.shape) 
print(x_train.shape)
cifar = reshape(x_train, [x_train.shape[0], x_train.shape[1]*x_train.shape[2]*x_train.shape[3]])
print(cifar.shape)
tsne = TSNE(n_components=2, verbose=1, random_state=123)
z = tsne.fit_transform(cifar)
df = pd.DataFrame()
df["y"] = y_train.mean()
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]

sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 10),
                data=df).set(title="CIFAR100 data T-SNE projection")
```
